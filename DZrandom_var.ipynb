```python
"""Условие:
Смоделировать игру против лотерейного автомата типа “777”. 
Игрок платит 1 руб., после чего выпадает случайное целое число, 
равномерно распределенное от 0 до 999. 
При некоторых значениях числа игрок получает выигрыше
777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.  
Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?"""
```

```python
attempt = 100000   # Количество попыток = Сумма проигранных денег (стоимость 1 попытки - 1 руб)

win = 0          # Сумма выйгранных денег
  
successes = 0    # Количество успехов


for i in range(attempt):
    number = np.random.randint(0, 999) # Случайное число от 0 до 999
    if number == 777:
        win += 200
        successes += 1
        
    elif number == 999:
        win += 100
        successes += 1
        
    elif number == 555:
        win += 50
        successes += 1
        
    elif number == 333:
        win += 15
        successes += 1
       
    elif number == 111:
        win += 10
        successes += 1
               
    elif  number%100 == 77:
        win += 5
        successes += 1        
       
    elif  number%10 == 7:
        win += 3
        successes += 1
        
    elif  number%100 == 0 and number != 0:
        win += 2
        successes += 1
        
    elif  number%10 == 0 and number != 0:
        win += 1
        successes += 1
        
    else:
        continue
        

p_successes = successes / attempt  # Вероятность выйгрыша
total = win - attempt

print (f'Количество выйгранных денег: {win}')
print (f'Количество потраченных денег: {attempt}')
print (f'Итого в кармане после игры: {total}')
print (f'Вероятность выигрыша: {p_successes}')
print (f'Игра не выгодна игроку')
```
